# 2. Person Class Using Inheritance

```python
# PERSON CLASS USING INHERITANCE

class Person:
    def __init__(self,first,last,email,DOB):
        self.first=first
        self.last=last
        self.email=email
        self.DOB=DOB

class Student(Person):
    def __init__(self,first,last,email,DOB,studentID,GPA):
        Person.__init__(self,first,last,email,DOB)
        self.studentID=studentID
        self.GPA=GPA

class Teacher(Person):
    def __init__(self,first,last,email,DOB,teacherID,ListOfStudents):
        Person.__init__(self,first,last,email,DOB)
        self.teacherID=teacherID
        self.ListOfStudents=ListOfStudents
        self.class_size=len(ListOfStudents)
    def show_student_info(self):
        print("Teacher Name:",self.first,self.last)
        for student in self.ListOfStudents:
            print("Student Name:",student.first,student.last,"Student Email:",student.email,"Student DOB:",student.DOB,"Student ID/GPA:",student.studentID,student.GPA)
    def add_students(self,students):
        for student in students:
            self.ListOfStudents.append(student)
    def remove_students(self,students):
        for student in students:
            if student in self.ListOfStudents:
                self.ListOfStudents.remove(student)

student1=Student("John","Smith","johnsmith@gmail.com","4-3-13",3674,3.5)
student2=Student("Bob","Roberts","robertsb@gmail.com","2-25-03",2829,3.7)
student3=Student("George","Brown","browng@gmail.com","5-3-11",4833,3.1)
teacher1=Teacher("Tom","Wilson","wilsont@gmail.com","7-30-82",7693,[student1,student2])
teacher2=Teacher("Tom","Wilson","wilsont@gmail.com","7-30-82",7693,[student1,student2])
teacher1.remove_students([student1])
teacher2.add_students([student3])
teacher1.show_student_info()
teacher2.show_student_info()
```

```python
# SHAPES CLASS USING INHERITANCE 

import pygame
import random
pygame.init()
screen=pygame.display.set_mode((600,600))
pygame.display.set_caption("Shapes")

class Shape:
    def __init__(self,size):
        self.x=random.randint(1,600)
        self.y=random.randint(1,600)
        self.color=(random.randint(0,255),random.randint(0,255),random.randint(0,255))
        self.size=size

class Circle(Shape):
    def __init__(self,size):
        Shape.__init__(self,size)
    def draw_circle(self):
        pygame.draw.circle(screen,self.color,(self.x,self.y),self.size)
        g=(self.color,self.x,self.y,self.size)
        return g
class Rectangle(Shape):
    def __init__(self,size):
        Shape.__init__(self,size)
    def draw_rectangle(self):
        pygame.draw.rect(screen,self.color,(self.x,self.y,self.size[0],self.size[1]))
        rect=(self.color,self.x,self.y,self.size[0],self.size[1])
        return rect

rectlist=[]
circlist=[]
sqrlist=[]
while True:

    screen.fill((0,0,0))
    
    for rect in rectlist:
        pygame.draw.rect(screen,rect[0],(rect[1],rect[2],rect[3],rect[4]))
    for circ in circlist:
        pygame.draw.circle(screen,circ[0],(circ[1],circ[2]),circ[3])
    for sqr in sqrlist:
        pygame.draw.rect(screen,sqr[0],(sqr[1],sqr[2],sqr[3],sqr[4]))

    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            pygame.quit()
            exit()
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_r:
                rectangle=Rectangle([20,10])
                r=rectangle.draw_rectangle()
                rectlist.append(r)
            if event.key==pygame.K_c:
                circle=Circle(20)
                c=circle.draw_circle()
                circlist.append(c)
            if event.key==pygame.K_s:
                square=Rectangle([15,15])
                s=square.draw_rectangle()
                sqrlist.append(s)
    pygame.display.update()
```

# 5. Circle Race*

```python
# CIRCLE RACE

import pygame
import random
pygame.init()
screen=pygame.display.set_mode((600,600))
pygame.display.set_caption("Moving Circles Class")
class Circle:
    def __init__(self,r,g,b,speed,y,x,moving_right,radius):
        self.color=(r,g,b)
        self.speed=speed
        self.y=y
        self.x=x
        self.moving=moving_right
        self.radius=radius
        self.lapcount=0
    def draw(self):
        pygame.draw.circle(screen,self.color,(self.x,self.y),self.radius)
    def move(self):
        if self.moving:
            self.x+=self.speed
        else:
            self.x-=self.speed
        
    def check_x(self):
        if self.x-self.radius>=600:
            self.moving=False
        if self.x+self.radius<0:
            self.moving=True
            self.lapcount+=1
    def check_lapcount(self):
        if self.lapcount>=10:
            self.x=30
circles=[]
for c in range(1,2,1):
    for n in range(1,11,1):
        circle=Circle(random.randint(0,255),random.randint(0,255),random.randint(0,255),random.randint(1,20),50*n,30,True,10)
        circles.append(circle)

clock=pygame.time.Clock()

start=False

while True:
    screen.fill((0,0,0))
    clock.tick(50)
    for d in circles:
        d.draw()
        if start:
            print("started")
            d.move()
            d.check_x()
            d.check_lapcount()
         
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            pygame.quit()
            exit()
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_SPACE:
                start=True
    pygame.display.update()
```

# 8. Shopping Cart

```python
# SHOPPING CART

class Shopping_List:
    def __init__(self,name):
        self.cart={}
        self.name=name
    def add(self,item,quantity):
        if item not in self.cart:
            self.cart[item]=quantity
        elif item in self.cart:
            self.cart[item]+=quantity
    def remove(self,item,remquantity):
        if item not in self.cart:
            print("The item is not in the cart.")
        elif item in self.cart:
            if remquantity>self.cart[item]:
                del self.cart[item]
            elif remquantity<=self.cart[item]:
                self.cart[item]-=remquantity
    def display_cart(self):
        print("Shop:",self.name,"\nShopping_List:",self.cart)
        
walmart=Shopping_List("Walmart")
walmart.add("Bananas",5)
walmart.display_cart()
walmart.remove("Bananas",3)
walmart.display_cart()
```

```python
# ALIEN INVASION

import random
import pygame
pygame.init()
screen=pygame.display.set_mode((1200,800))
pygame.display.set_caption("Space Invaders")


class Character:
    def __init__(self,x,y,length,width):
        self.x=x
        self.y=y
        self.image=random.choice(["Object Oriented Programming/Alien2.png","Object Oriented Programming/Alien3.png","Object Oriented Programming/Alien1.png"])
        self.length=length
        self.width=width
    def draw(self):
        image=pygame.image.load(self.image)
        image=pygame.transform.scale(image,(self.length,self.width))
        screen.blit(image,(self.x,self.y))

class Alien(Character):
    def move_alien(self,forwards,vel):
        if forwards:
            self.x+=vel
        else:
            self.x-=vel

class Ship(Character):
    def __init__(self,x,y,length,width):
        Character.__init__(self,x,y,length,width)
        self.image=("Object Oriented Programming/Spaceship.png")
    def move_ship_right(self):
        self.x+=10
        if self.x>=1200-self.length:
            self.x=1200-self.length
        if self.x<=0:
            self.x=0
    def move_ship_left(self):
        self.x-=10
        if self.x>=1200-self.length:
            self.x=1200-self.length
        if self.x<=0:
            self.x=0
class Bullet(Character):
    def __init__(self,x,y,length,width):
        Character.__init__(self,x,y,length,width)
        self.image=("Object Oriented Programming/Bullet.png")
        self.rect=pygame.Rect((self.x,self.y),(self.length,self.width))
    def move_bullet(self):
        self.y-=50

       


alienlist=[]
def init(rows,collumns):
    for r in range(1,rows,1):
        for c in range(1,collumns,1):
            alien=Alien(c*150,r*65,70,50)
            alien.draw()
            alienlist.append(alien)
r=6
c=6
level=0
bg=pygame.image.load("Object Oriented Programming/Nebula.png")   
bg=pygame.transform.scale(bg,(1200,800))
playership=Ship(550,650,100,100)
clock=pygame.time.Clock()
direction=True
right,left=False,False
bulletlist=[]

v=10

def show_text(msg,x,y,color,size):
    fontobj=pygame.font.SysFont("freesans",size)
    msgobj=fontobj.render(msg,False,color)
    screen.blit(msgobj,(x,y))

while True:
    clock.tick(15)
    screen.blit(bg,(0,0))
    show_text("Level: "+str(level),50,40,(100,0,100),25)
    playership.draw()
    if right:
        playership.move_ship_right()
    if left:
        playership.move_ship_left()
    
    for bullet in bulletlist:
        bullet.move_bullet()
        if bullet.y<=-80:
            bulletlist.remove(bullet)
        for alien in alienlist:
            b=pygame.Rect((bullet.x,bullet.y),(bullet.length,bullet.width))
            a=pygame.Rect((alien.x,alien.y),(alien.length,alien.width))
            if b.colliderect(a):
                print("collision detected")
                if bullet in bulletlist:
                    bulletlist.remove(bullet)
                if alien in alienlist:
                    alienlist.remove(alien)
        
        bullet.draw()
 
    for alien in alienlist:
    
        a=pygame.Rect((alien.x,alien.y),(alien.length,alien.width))
        s=pygame.Rect((playership.x,playership.y),(playership.length,playership.width))
        alien.draw()
        alien.move_alien(forwards=direction,vel=v)
        if alien.x<=0 or alien.x+70>=1200:
            if direction:
                direction=False
            elif not direction:
                direction=True
            for a in alienlist:
                a.y+=65
        a=pygame.Rect((alien.x,alien.y),(alien.length,alien.width))
        s=pygame.Rect((playership.x,playership.y),(playership.length,playership.width))
        if s.colliderect(a):
           
            r=6
            c=6
            level=0
            playership=Ship(550,650,100,100)
            direction=True
            right,left=False,False
            bulletlist=[]
            alienlist=[]
            v=10
            init(rows=r,collumns=c)
                              
    
    if len(alienlist)==0:
        v+=1
        level+=1
        r+=0
        init(rows=r,collumns=c)
    for event in pygame.event.get():
        if event.type==pygame.QUIT:
            pygame.quit()
            exit()
        if event.type==pygame.KEYDOWN:
            if event.key==pygame.K_LEFT:
                left=True
                right=False
            if event.key==pygame.K_RIGHT:
                right=True
                left=False
            if event.key==pygame.K_SPACE:
                bullet=Bullet(playership.x+10,610,80,80)
                bulletlist.append(bullet)
                bullet.draw()
        if event.type==pygame.KEYUP:
            if event.key==pygame.K_LEFT:
                left,right=False,False
            if event.key==pygame.K_RIGHT:
                right,left=False,False
    pygame.display.update()
```
